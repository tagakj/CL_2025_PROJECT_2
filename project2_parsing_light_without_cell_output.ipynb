```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import time
import re
```

```python
page = requests.get('https://mamadeti.ru/qa/')
print(page)
```

```python
# генерим ссылки на страницы с q&a (всего 411 страниц), тексты обращений располагаются прямо на этих страницах
urls = ['https://mamadeti.ru/qa/']
for i in range(2, 412):
  urls.append("https://mamadeti.ru/qa/" + "?nav-qa=page-" + str(i))
print(len(urls), urls)
```

```python
# ЗАПУСК. УБРАТЬ ПРИНТ data!!!
data = []
for url in urls:
  try:
    page = requests.get(url)
    time.sleep(3)
    soup = BeautifulSoup(page.text, features = "html.parser")
    for i in soup.find_all('div', {'class': 'qa-card'}):
      data.append([i.find('span', {'class': 'qa-card__section'}).text, i.find('span', {'class': 'qa-card__date'}).text, i.find('a').text, i.find('div', {'class': 'qa-card__text text'}).text])
  except requests.exceptions.RequestException as e:
    print(f"Ошибка на {url}: {e}")
print(len(data), data)
```

```python
df_qa = pd.DataFrame(data).rename(columns={0: "category", 1: "date", 2: "question_theme", 3: "question_text"})
df_qa
```

```python
df_qa.to_csv('qa_table.csv', encoding='utf-8')
df_qa.to_excel('qa_table.xlsx')
```
